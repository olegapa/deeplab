```python
import os
import json
import matplotlib.pyplot as plt
INPUT_DATA = 'jsons_hist'

files_in_directory = [
    os.path.join(INPUT_DATA, f)
    for f in os.listdir(INPUT_DATA)
    if (
        os.path.isfile(os.path.join(INPUT_DATA, f))
    )
]
files_in_directory = [
    file for file in files_in_directory if file.split('.')[-1] == 'json'
]
```

```python
for file in files_in_directory:
    with open(file, 'r', encoding='utf-8') as input_json:
        json_data = json.load(input_json)
    for label, values in json_data.items():
        f, ax = plt.subplots(figsize=(15,8))
        k = list(values.keys())
        v = list(values.values())
        new_k = list()
        new_v = list()
        for i in range (25):
#             print(float(k[4*i + 3]))
            new_k.append(float(k[4*i + 3]))
            new_v.append(v[4*i] + v[4*i + 1] + v[4*i + 2] + v[4*i + 3])
#             print(new_k[i])
#             print(new_v[i])
        plt.bar(new_k, new_v, label=str(label), width=0.03)
        plt.show()
```
