```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
INPUT_DATA = 'json_class_hist'
n_cls=8
files_in_directory = [
    os.path.join(INPUT_DATA, f)
    for f in os.listdir(INPUT_DATA)
    if (
        os.path.isfile(os.path.join(INPUT_DATA, f))
    )
]
files_in_directory = [
    file for file in files_in_directory if file.split('.')[-1] == 'json'
]
hist=dict()
for  i in range(n_cls+1):
    hist[i] = dict()
    for j in np.arange(0, 1, 0.01):
        hist[i][j]=0
```

```python
#old case
for file in files_in_directory:
    with open(file, 'r', encoding='utf-8') as input_json:
        json_data = json.load(input_json)
    for item in json_data['files']:
        for chain in item['file_chains']:
            for frame in chain['chain_markups']:
                if frame['markup_path'].get('polygons', None):
                    for cls, pol in frame['markup_path']['polygons'].items():
                        print(pol["score"] - pol["score"]%0.01)
                        hist[int(cls)][pol["score"] - pol["score"]%0.01] += 1
```

```python
for label, values in hist.items():
    f, ax = plt.subplots(figsize=(15,8))
    k = list(values.keys())
    v = list(values.values())
    new_k = list()
    new_v = list()
    for i in range (50):
#           print(float(k[4*i + 3]))
        new_k.append(float(k[2*i + 1]))
        new_v.append(v[2*i] + v[2*i + 1])
#             print(new_k[i])
#             print(new_v[i])
#         print(new_k)
#         print(new_v)
    plt.bar(new_k, new_v, width=0.015)
    plt.show()
```
